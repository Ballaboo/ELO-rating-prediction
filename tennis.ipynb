```python
# Import depencies in venv
import pandas as pd
import numpy as np
import os
```

```python
# Expand csv path to user 
csv_path = os.path.expanduser('~/Project/tennis/tennis_atp/')

# List to hole append
csv_combine = []

# Loop through 2000-2024
for year in range(2000, 2024):
    file_path = os.path.join(csv_path, f'atp_matches_{year}.csv')
    df = pd.read_csv(file_path)  # This should now work correctly
    csv_combine.append(df)

# Concat akk dataframes
df_concat = pd.concat(csv_combine, ignore_index=True)
```
